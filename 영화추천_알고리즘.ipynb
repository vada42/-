```python
import pandas as pd
import numpy as npK
from sklearn.neighbors import KNeighborsClassifier
movie=pd.read_csv('https://raw.githubusercontent.com/vada42/vvv/main/movie_data.csv')
movie
```

```python
movie_title=movie['title']
movie_title=list(movie_title)
print(movie_title)
```

```python
print("action, drama, comedy, fantasy, romance, crime, horror, musical, thriller, sf, documentary, mystery")
user_select1= input('  : ')
user_select2= input('  : ')
user_select3= input('  : ')
```

```python
movie_input = movie[[user_select1,user_select2,user_select3]].to_numpy()

print(movie_input)
```

```python
user_select1_score=int(input(user_select1 + " "+"score: "))
user_select2_score=int(input(user_select2 + " "+"score: "))
user_select3_score=int(input(user_select3 + " "+"score: "))
user = [user_select1_score,user_select2_score,user_select3_score]
print(user)
```

```python
kn = KNeighborsClassifier(n_neighbors=1)
kn.fit(movie_input, movie_title)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(movie_input[:,0], movie_input[:,1], movie_input[:,2],marker='o')
ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_zlim(0, 100)
ax.scatter(user[0], user[1], user[2],marker='^')
ax.set_xlabel(user_select1)
ax.set_ylabel(user_select2)
ax.set_zlabel(user_select3)
plt.show()
distances, indexes = kn.kneighbors([user])
user_suggestion=kn.predict([user])
print(user_suggestion)
print(distances)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(movie_input)
movie_scaled = ss.transform(movie_input)
```

```python
kn.fit(movie_scaled, movie_title)
print(kn.score(movie_scaled, movie_title))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(movie_scaled[:,0], movie_scaled[:,1], movie_scaled[:,2],marker='o')
ax.set_xlabel(user_select1)
ax.set_ylabel(user_select2)
ax.set_zlabel(user_select3)
plt.show()
```

```python
mean = np.mean(movie_input, axis=0)
std=np.std(movie_input, axis=0)
```

```python
print(mean, std)
movie_scaled = (movie_input - mean)/std
```

```python
kn.fit(movie_scaled, movie_title)
```

```python
user =([user_select1_score,user_select2_score,user_select3_score]-mean)/std
```
